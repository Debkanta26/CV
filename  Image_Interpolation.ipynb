```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. LOAD YOUR UPLOADED IMAGE
# Use the 'r' prefix for Windows paths to avoid errors
path = r"C:\Users\KIIT0001\Desktop\Program\Python Projects\input.jpeg"
img = cv2.imread(path)

if img is None:
    print("Error: Could not find the image. Please check the file path.")
else:
    # Convert BGR to RGB for correct colors in Jupyter
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Define how much you want to resize (e.g., 2x larger or a specific size)
    # For training your Keras model, you used (60, 60)
    target_size = (60, 60) 

    # 2. APPLY INTERPOLATION METHODS
    # Nearest Neighbor (Fast, preserves edges but blocky)
    res_nearest = cv2.resize(img_rgb, target_size, interpolation=cv2.INTER_NEAREST)
    
    # Bilinear (Standard for general resizing)
    res_bilinear = cv2.resize(img_rgb, target_size, interpolation=cv2.INTER_LINEAR)
    
    # Area (The BEST for shrinking images/creating datasets)
    res_area = cv2.resize(img_rgb, target_size, interpolation=cv2.INTER_AREA)
    
    # Bicubic (High quality for enlarging images)
    res_bicubic = cv2.resize(img_rgb, target_size, interpolation=cv2.INTER_CUBIC)

    # 3. DISPLAY THE OUTPUTS
    titles = ['Original', 'Nearest', 'Bilinear', 'Area (Best for Shrinking)', 'Bicubic']
    images = [img_rgb, res_nearest, res_bilinear, res_area, res_bicubic]

    plt.figure(figsize=(20, 10))
    for i in range(5):
        plt.subplot(1, 5, i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```
